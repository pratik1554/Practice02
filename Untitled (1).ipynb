```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

from sklearn.model_selection import RandomizedSearchCV, GridSearchCV

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import plot_roc_curve
```

```python
data= pd.read_csv('Ionosphere.csv')
data.head()
```

```python
data.drop('V2',axis=1,inplace=True)
```

```python
data.describe()
```

```python
data.corr()
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(data.corr(),annot=True,cmap='rainbow');
```

```python
X=data.drop('Class',axis=1)
y= data['Class']
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,stratify=y,random_state=25)
```

```python
models={'Logistic Regression': LogisticRegression(),
        'Decision tree':DecisionTreeClassifier(),
        'KNN':KNeighborsClassifier(n_neighbors=2),
        'Gaussian':GaussianNB(),
        'XGB':XGBClassifier(),
        'svc':SVC(),
       'Random_forest': RandomForestClassifier()}
```

```python
model_score={}

for name,model in models.items():
    model.fit(X_train,y_train)
    model_score[name]=model.score(X_test,y_test)
```

```python
model_score
```

```python
model_train={}

for name,model in models.items():
    model.fit(X_train,y_train)
    model_train[name]=model.score(X_train,y_train)
```

```python
model_train
```

```python
model_compare = pd.DataFrame(model_score, index=["accuracy"])
model_compare.T.plot.bar();
```

```python
# Cross Validation
```

```python
lr_2= LogisticRegression(C=32.90344562312671, solver= 'liblinear')
```

```python
cv_acc = cross_val_score(lr_2,
                         X,
                         y,
                         cv=5,
                         scoring="accuracy")
```

```python
cv_acc
```

```python
cv_acc = np.mean(cv_acc)
cv_acc
```

```python
cv_precision = cross_val_score(lr_2,
                         X,
                         y,
                         cv=5,
                         scoring="precision")
cv_precision=np.mean(cv_precision)
cv_precision
```

```python
cv_recall = cross_val_score(lr_2,
                         X,
                         y,
                         cv=5,
                         scoring="recall")
cv_recall = np.mean(cv_recall)
cv_recall
```

```python
cv_f1 = cross_val_score(lr_2,
                         X,
                         y,
                         cv=5,
                         scoring="f1")
cv_f1 = np.mean(cv_f1)
cv_f1
```

```python
cv_metrics = pd.DataFrame({"Accuracy": cv_acc,
                           "Precision": cv_precision,
                           "Recall": cv_recall,
                           "F1": cv_f1},
                          index=[0])

cv_metrics.T.plot.bar(title="Cross-validated classification metrics",
                      legend=False);
```

```python
RFC=RandomForestClassifier(random_state=25)
RFC.fit(X_train,y_train)
y_predict=RFC.predict(X_test)
cm=confusion_matrix(y_test,y_predict)
print('RFC Score=',RFC.score(X_test,y_test)*100)
print('\n')
print('confusion Matrix is:',cm,sep='\n')
print('\n')
from sklearn import metrics
probs=RFC.predict_proba(X_test)
prob_positive =probs[:,1]
fpr,tpr,threshold=metrics.roc_curve(y_test,prob_positive)
roc_auc=metrics.auc(fpr,tpr)
print('Area under the curve:',roc_auc)
print('\n')
plt.title('Recevier Operating Characteristic')
plt.plot(fpr,tpr,'orange',label='AUC=%0.2f' % roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],color='darkblue',linestyle='--')
plt.ylabel('True Positive rate')
plt.xlabel('False Positive rate')
plt.show()

```
